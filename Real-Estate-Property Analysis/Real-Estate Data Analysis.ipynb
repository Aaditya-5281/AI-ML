```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
properties = pd.read_csv('properties.csv')
```

```python
# Data Visualizations

# 1. Distribution of Sale Prices
plt.figure(figsize=(10, 6))
sns.histplot(data=properties, x='SalePrice', bins=30, kde=True)
plt.title('Distribution of Sale Prices', fontsize=16)
plt.xlabel('Sale Price', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

```python
# 2. Average Sale Price by Location
location_price = properties.groupby('Location')['SalePrice'].mean().reset_index()
plt.figure(figsize=(8, 6))
sns.barplot(x='Location', y='SalePrice', data=location_price)
plt.title('Average Sale Price by Location', fontsize=16)
plt.xlabel('Location', fontsize=14)
plt.ylabel('Average Sale Price', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Relationship between Size and Sale Price
plt.figure(figsize=(10, 6))
sns.scatterplot(data=properties, x='Size_sqft', y='SalePrice')
plt.title('Relationship between Size and Sale Price', fontsize=16)
plt.xlabel('Size (sq ft)', fontsize=14)
plt.ylabel('Sale Price', fontsize=14)
plt.show()
```

```python
# 4. Box Plot of Sale Prices by Location
plt.figure(figsize=(10, 6))
sns.boxplot(data=properties, x='Location', y='SalePrice')
plt.title('Box Plot of Sale Prices by Location', fontsize=16)
plt.xlabel('Location', fontsize=14)
plt.ylabel('Sale Price', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```
